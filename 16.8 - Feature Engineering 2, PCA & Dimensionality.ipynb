1. Create a new score variable from the weighted averages of all score variables in the datasets. Notice that the number of students in the 4th grade isn't the same as the number of students in the 8th grade. So, you should appropriately weigh the scores!


```python
def weighted_average(df):
    df['WEIGHTED_AVERAGE'] = ''
    for i, row in df.iterrows():
        total_students = df.iloc[i]['GRADES_4_G'] + df.iloc[i]['GRADES_8_G']
        grade_4_weight = df.iloc[i]['GRADES_4_G'] / total_students
        grade_8_weight = df.iloc[i]['GRADES_8_G'] / total_students
        grade_4_average = ((df.iloc[i]['AVG_MATH_4_SCORE'] + df.iloc[i]['AVG_READING_4_SCORE']) / 2)
        grade_8_average = ((df.iloc[i]['AVG_MATH_8_SCORE'] + df.iloc[i]['AVG_READING_8_SCORE']) / 2)
        weighted_average = grade_4_average * grade_4_weight + grade_8_average * grade_8_weight
        df.iloc[i, df.columns.get_loc('WEIGHTED_AVERAGE')] = weighted_average
    
wa_usedu = weighted_average(usedu)
```
